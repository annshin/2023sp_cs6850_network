```python
def get_sorted_meme_scores_for_year(discipline_suffix, year):
    meme_score_path = os.path.join(CACHE_PATH, f'year_{year}_meme_scores_{discipline_suffix}.pkl')
    with open(meme_score_path, 'rb') as f:
        data = pickle.load(f)
    
    meme_score_pairs = []
    for meme, score_dict in data.items():
        meme_score_pairs.append((meme, score_dict['meme_score']))
    meme_score_pairs = sorted(meme_score_pairs, reverse=True, key=lambda x: x[1])
    return meme_score_pairs
```

```python
meme_scores = get_sorted_meme_scores_for_year('ai', 2022)
for meme, score in meme_scores[:50]:
    print(meme, score)
```

```python
key_years = [2002, 2007, 2012, 2017, 2022]
yearly_differences = {discipline : [] for discipline in discipline_to_top_15.keys()}
for discipline, year_dict in discipline_to_top_15.items():
    for i in range(4):
        curr_year = key_years[i]
        next_year = key_years[i+1]
        
        next_memes = set(year_dict[next_year])
        curr_memes = set(year_dict[curr_year])
        
        new_meme_count = 0
        for meme in next_memes:
            if meme not in curr_memes:
                new_meme_count += 1
                
        yearly_differences[discipline].append(new_meme_count)
```

```python
def get_meme_score(meme_score, meme):
    for curr_meme, score in meme_score:
        if meme == curr_meme:
            return score
    return 0

term_to_meme_score = {term : [] for term in terms}
for year in range(1991, 2023):
    meme_score, _ = get_sorted_meme_scores_for_year(meme_scores, year)
    for term in terms:
        term_to_meme_score[term].append(get_meme_score(meme_score, term))
```
